Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Loading and Preprocessing

```python
#load dataframe
df = pd.read_csv('restaurants.csv')
df.head()
```

```python
#create a refined dataframe
df1 = df[['Restaurant ID','Restaurant Name','Cuisines','Price range','Aggregate rating','Votes']]
df1
```

```python
#handle missing values
df1.isna().sum()
```

```python
df1 = df1.dropna()
df1.isna().sum()
```

```python
df1.duplicated().sum()
```

```python
df1['Restaurant Name'].duplicated().sum()
```

```python
df1['Restaurant Name'].value_counts()
```

```python
#sorting the restaurants by name and rating
df1 = df1.sort_values(by=['Restaurant Name','Aggregate rating'],ascending=False)
df1.head()
```

```python
df1[df1["Restaurant Name"]=="Cafe Coffee Day"].head()
```

```python
#removing duplicate entries of same restaurant name
df1 = df1.drop_duplicates('Restaurant Name',keep='first')
df1
```

```python
df1['Restaurant Name'].value_counts()
```

```python
df1 = df1[df1['Aggregate rating']>3.9]
df1
```

```python
#splitting cuisines into list
df1['Cuisines'] = df1['Cuisines'].str.split(',')
df1
```

```python
df1 = df1.explode('Cuisines')
df1
```

```python
df1['Cuisines'].value_counts()
```

```python
restoXcuisines = pd.crosstab(df1['Restaurant Name'], df1['Cuisines'])
```

```python
restoXcuisines
```

```python
df1['Restaurant Name'].sample(20, random_state=194)
```

```python
from sklearn.metrics import jaccard_score
```

```python
print(jaccard_score(restoXcuisines.loc["Olive Bistro"].values,
                    restoXcuisines.loc["Rose Cafe"].values))
```

```python
from scipy.spatial.distance import pdist, squareform
```

```python
jaccardDist = pdist(restoXcuisines.values, metric='jaccard')
jaccardMatrix = squareform(jaccardDist)
jaccardSim = 1 - jaccardMatrix
dfJaccard = pd.DataFrame(
    jaccardSim,
    index=restoXcuisines.index,
    columns=restoXcuisines.index)

dfJaccard
```

```python
df1['Restaurant Name'].sample(20)
```

```python
resto = 'Ooma'

sim = dfJaccard.loc[resto].sort_values(ascending=False)
sim = pd.DataFrame({'Restaurant Name': sim.index, 'simScore': sim.values})
sim = sim[(sim['Restaurant Name']!= resto) & (sim['simScore']>=0.7)].head(5)
RestoRec = pd.merge(sim,df1[['Restaurant Name','Aggregate rating']],how='inner',on='Restaurant Name')
FinalRestoRec = RestoRec.sort_values('Aggregate rating',ascending=False).drop_duplicates('Restaurant Name',keep='first')
```
